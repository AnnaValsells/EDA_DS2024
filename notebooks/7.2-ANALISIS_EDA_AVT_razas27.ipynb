```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
sns.set_style("white")
```

```python
df_agility = pd.read_csv("./data/6_DATA_AGILITY.csv")
```

```python
df_agility.info()
```

### Hipotesis 1.1: Los perros de raza border collie no ganan más que los perros de otras razas.

```python
df_agility.groupby(df_agility["Breed"] == "Border Collie")["Classification"].value_counts()
```

```python
# mas de 100 pistas realizadas:
razas_27 = df_agility["Breed"].value_counts()[:27]
razas_27
```

```python
# de 32611 pistas, 11314 son realizadas por Border Collies.
sns.barplot(y= razas_27.index, x= razas_27.values, hue= razas_27.values, legend=False, palette= "tab20b")
plt.title("Número de pistas relalizadas")
plt.ylabel("")
plt.xlabel("32611 pistas analizadas")
plt.grid(False)
plt.xlim(0,12500)
plt.xticks([])
sns.despine(left=True, bottom=True)

for i, valor in enumerate(razas_27):
    plt.text(valor, i, str(valor), ha='left', va='center')


# graf_total_razas.savefig("pistas_totales_por_razas.png") --> Guardado a mano
```

```python
razas_27_podium = pd.DataFrame(razas_27)
dic_podium = {
    "Primero":[],
    "Segundo":[],
    "Tercero":[]
}
for raza in razas_27_podium.index:
    podium_x_raza = df_agility.groupby(df_agility["Breed"] == raza)["Classification"].value_counts()
    for a,b in podium_x_raza.items():
        if a == (True, '1º'):
            dic_podium["Primero"].append(b)
        if a == (True, '2º'):
            dic_podium["Segundo"].append(b)
        if a == (True, '3º'):
            dic_podium["Tercero"].append(b)

# Los chiuauas no tienen podim en bronce (3ª)
df_agility.groupby(df_agility["Breed"] == "Chihuahua")["Classification"].value_counts()
```

```python
dic_podium["Tercero"] = np.insert(dic_podium["Tercero"], 18, 0)
```

```python
razas_27_podium["Primero"] = dic_podium["Primero"]
razas_27_podium["Segundo"] = dic_podium["Segundo"]
razas_27_podium["Tercero"] = dic_podium["Tercero"]
razas_27_podium["Suma_Podiums"] = razas_27_podium["Primero"] + razas_27_podium["Segundo"] + razas_27_podium["Tercero"]
```

```python
razas_27_podium = razas_27_podium.sort_values(by="Suma_Podiums", ascending=False)
```

```python
razas_27_podium["%Absoluto"] = razas_27_podium["Suma_Podiums"] / 32611 *100 
razas_27_podium["%Relativo"] = razas_27_podium["Suma_Podiums"] / razas_27_podium["count"] *100
```

```python
razas_27_podium
```

```python
# En terminos relativos
podium_relativo = razas_27_podium.sort_values(by="%Relativo", ascending=False)

sns.barplot(y= podium_relativo.index, x= podium_relativo["%Relativo"], hue= podium_relativo["%Relativo"], legend=False, palette= "tab20b")
plt.title("% de Perros que llegan a hacer podio")
plt.ylabel("")
plt.xlabel("");
```

```python
# En terminos absolutos
podium_absoluto = razas_27_podium.sort_values(by="%Absoluto", ascending=False)

fig = plt.plot(figsize= (100,50))
sns.barplot(y= podium_absoluto.index, x= podium_absoluto["%Absoluto"], hue= podium_absoluto["%Absoluto"], legend=False, palette= "tab20b")
fig
```

```python
dic_elim = {
    "pistas_a_0" : [],
    "eliminados" : []
}

for raza in razas_27_podium.index:
    elim_x_raza = df_agility.groupby(df_agility["Breed"] == raza)["Eliminados"].value_counts()
    for a,b in elim_x_raza.items():
        if a == (True, 'EXC_0'):
            dic_elim["pistas_a_0"].append(b)
        if a == (True, 'ELIM'):
            dic_elim["eliminados"].append(b)
```

```python
razas_27_podium["pistas_a_0"] = dic_elim["pistas_a_0"]
razas_27_podium["eliminados"] = dic_elim["eliminados"]
```

```python
razas_27_podium["%Pisats_a_0"] =  razas_27_podium["pistas_a_0"] / razas_27_podium["count"] *100
razas_27_podium["%eliminados"] =  razas_27_podium["eliminados"] / razas_27_podium["count"] *100
```

```python
#SEGUIR ESTUDIANDO: Primeros - Suma podiums - pistas a 0 // Eliminados

a_cero_relativo = razas_27_podium.sort_values(by="%Pisats_a_0", ascending=False)

fig = plt.plot(figsize= (100,50))
sns.barplot(y= a_cero_relativo.index, x= a_cero_relativo["%Pisats_a_0"], hue= a_cero_relativo["%Pisats_a_0"], legend=False, palette= "tab20b")
plt.xlim(0,80)
plt.title("% de Perros que tienen pistas a cero")
plt.ylabel("")
plt.xlabel("")
fig
```

```python
eliminados_relativo = razas_27_podium.sort_values(by="%eliminados", ascending=True)

fig = plt.plot(figsize= (100,50))
sns.barplot(y= eliminados_relativo.index, x= eliminados_relativo["%eliminados"], hue= eliminados_relativo["%eliminados"], legend=False, palette= "tab20b")
plt.xlim(0,80)
plt.title("% de Perros que tienen eliminados en pista")
plt.ylabel("")
plt.xlabel("")
fig
```
